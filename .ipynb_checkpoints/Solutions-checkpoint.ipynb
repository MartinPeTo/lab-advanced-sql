```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="/home/martin/Ironhack - Labs/Ironhack Data Analytics.json"
from google.cloud import bigquery
client = bigquery.Client()
```

```python
step1="""
SELECT
    title.title_id,
    authors.au_id,
    (title.price * sales.qty * title.royalty / 100 * title_author.royaltyper / 100) as Royalty
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
INNER JOIN
    `ironhack-data-analytics.publications.sales` sales
ON
    title_author.title_id = sales.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.titles` title
ON
    sales.title_id = title.title_id

ORDER BY
    Royalty DESC
"""
```

```python
query_job=client.query(query=step1)
df=query_job.to_dataframe()
df
```

```python
step2="""
SELECT
    title.title_id,
    authors.au_id,
    SUM(title.price * sales.qty * title.royalty / 100 * title_author.royaltyper / 100) as Royalty
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
INNER JOIN
    `ironhack-data-analytics.publications.sales` sales
ON
    title_author.title_id = sales.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.titles` title
ON
    sales.title_id = title.title_id
    
GROUP BY

    title.title_id,
    authors.au_id

ORDER BY
    Royalty DESC
"""
```

```python
query_job=client.query(query=step2)
df=query_job.to_dataframe()
df
```

```python
step3="""
SELECT
    authors.au_id,
    SUM(title.price * sales.qty * title.royalty / 100 * title_author.royaltyper / 100) as Profit
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
INNER JOIN
    `ironhack-data-analytics.publications.sales` sales
ON
    title_author.title_id = sales.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.titles` title
ON
    sales.title_id = title.title_id
    
GROUP BY
    authors.au_id

ORDER BY
    Profit DESC
    
LIMIT
    3
    
"""
```

```python
query_job=client.query(query=step3)
df=query_job.to_dataframe()
df
```
